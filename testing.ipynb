```python
import pandas as pd

tennis_data_2021 = pd.read_csv("data/atp_matches_2021.csv", delimiter=',')
```

```python
tennisdata_2021.head()
```

```python
ao_2021 = tennis_data_2021[tennis_data_2021['tourney_name'] == 'Australian Open']
ao_2021.head()
```

```python
ao_2021.columns
```

```python
player_names = set(ao_2021['winner_id'].unique()) | set(ao_2021['loser_id'].unique())
player_names = sorted(player_names)
player_dtype = pd.CategoricalDtype(categories=player_names)
player_dtype
```

```python
from collections import defaultdict
players = defaultdict(dict)

for player in (player_names):
    matches = ao_2021[ao_2021[['winner_id', 'loser_id']].isin([player]).any(axis=1)]

    players[player] = matches[['w_ace', 'l_ace']]



```

```python
player_stats = pd.DataFrame.from_dict(players, orient='index')
```
